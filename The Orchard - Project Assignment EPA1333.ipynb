## Creating L-systems

Examples from Wikipedia: https://en.wikipedia.org/wiki/L-system

### Example 1: Algae
#### _Lindenmayer's original L-system for modelling the growth of algae._

The code beneath will write an L system with the following rules:

| Parameters | Value |
| --- | --- |
| Variables | A, B |
| Constants | None |
| Axiom | A |
| Rules | (A → AB), (B → A) |

To get some inspitarion: https://runestone.academy/ns/books/published/thinkcspy/Strings/TurtlesandStringsandLSystems.html

```python
# Create L-System with A, A->B and B -> AB
axiom = "A"
total_rounds = 7

results = [axiom]   # The result of round N can be printed with round[N]

for previous_round in range(total_rounds):
    string = ''
    for symbol in results[previous_round]:
        if symbol == 'A':   # First rule
            string += "AB"
        if symbol == 'B':   # Second rule
            string += "A"
    results += [string]

print(results)
```

### Example 2: Fractal (binary) tree

The code beneath will write an L system with the following rules:

| Parameters | Value |
| --- | --- |
| Variables | 0, 1 |
| Constants | "\[", "\]" |
| Axiom | 0 |
| Rules | (1 → 11), (0 → 1\[0\]0) |

> The shape is built by recursively feeding the axiom through the production rules. Each character of the input string is checked against the rule list to determine which character or string to replace it with in the output string. In this example, a '1' in the input string becomes '11' in the output string, while '\[' remains the same. Applying this to the axiom of '0', we get:

axiom:	0

1st recursion:	1[0]0

2nd recursion:	11[1[0]0]1[0]0

3rd recursion:	1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0

…

We can see that this string quickly grows in size and complexity. This string can be drawn as an image by using turtle graphics, where each symbol is assigned a graphical operation for the turtle to perform. For example, in the sample above, the turtle may be given the following instructions:

* 0: draw a line segment ending in a leaf
* 1: draw a line segment
* [ : push position and angle, turn left 45 degrees
* ] : pop position and angle, turn right 45 degrees

Inspiration from: https://runestone.academy/ns/books/published/thinkcspy/Strings/TurtlesandStringsandLSystems.html

```python
axiom = "0"
total_rounds = 3

results = [axiom]   # The result of round N can be printed with round[N]

for previous_round in range(total_rounds):
    string = ''
    for symbol in results[previous_round]:
        if symbol == '0':   # First rule
            string += "1[0]0"
        if symbol == '1':   # Second rule
            string += "11"
            
        if symbol == '[':   # First constant
            string += "["
        if symbol == ']':   # First constant
            string += "]"
    results += [string]

print(results)
```

### Creating an alogithm

The input is given in the tables. The axiom is just a variable and the variables are inherit to the system: these don't change the algorithm. What does are the constants and the rules.

Constants (1) are straightforward: if one is detected, it is kept the same.
The rules (2) are simple as well, input X is transformed to output Y.

1. Constants can be managed in two ways: testing if it exists and pasting it into the new string, or just copying if it is not in any rules. The first one takes an extra step in testing all the time, while for "just copying" only the axiom needs to be tested. Also rules can be written incorrectly, so they need to be tested aswell for "allowed characters" in the beginning. Therefore the code needs two parts, one that tests the input of the axiom and the rules, and the algorithm itself. Or you could avoind the testing by assuming that every symbol as rule output or axiom that is not in rule key, is a constant.
   * An upgrade would be a test if the axiom is in the rules, so that the user knows that this may be a whong input. Then the user should be prompted with a y/n to continue with this variable, with n resulting of the giving of a new variable? 
2. Rules should be a dictionary with one symbol being the key for what needs to be pasted in.

```python
# Second example with rules as a dict
axiom = "0"
rules = {'0': '1[0]0',
         '1': '11'}
total_rounds = 8

results = [axiom]   # The result of round N can be printed with round[N]

for previous_round in range(total_rounds):
    string = ''
    for symbol in results[previous_round]:
        try:
            string += rules[symbol]   # Implement the rule if it exists
        except:
            string += symbol          # Else treat the symbol as a constant
    results += [string]

print(results[8])
```

In a function:

```python
def create_L_system(axiom, rules, total_rounds):
    results = [axiom]   # The result of round N can be printed with round[N]

    for previous_round in range(total_rounds):
        string = ''
        for symbol in results[previous_round]:
            try:
                string += rules[symbol]   # Implement the rule if it exists
            except:
                string += symbol          # Else treat the symbol as a constant
        results += [string]

    return results
```

## Drawing L-systems using turtles

Given the previously explained L-system, the following needs to happen in the first round:
* The turtle needs to be pointed upwards
* Go forward the amount of 1's intil place "P"
* Then draw the first branch from P, following the codes until it hits 0.
* When it hits 0, it needs top return to the previous turn (P).
* Then start the second branch from P

When one branch also has multiple branches, P ccan be the last point.

* 0: draw a line segment ending in a leaf
* 1: draw a line segment
* \[ : push position and angle, turn left 45 degrees
* \] : pop position and angle, turn right 45 degrees

for the pushing and popping of position, recursion can be used.
